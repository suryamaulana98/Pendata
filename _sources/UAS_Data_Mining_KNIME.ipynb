## 2. Pendahuluan

Perkembangan data mining dalam bidang pendidikan, yang sering dikenal dengan istilah *Educational Data Mining* (EDM), telah menjadi salah satu inovasi terpenting dalam upaya meningkatkan kualitas sistem pembelajaran. Data mining memungkinkan pengolahan data berskala besar yang dikumpulkan oleh institusi akademik untuk diekstraksi menjadi informasi dan pola pengetahuan yang berharga. Teknik ini memfasilitasi penemuan hubungan tersembunyi antara perilaku mahasiswa dan hasil belajar mereka.

Mengetahui faktor-faktor yang mempengaruhi prestasi akademik mahasiswa adalah hal yang sangat penting. Performa akademik seringkali bukan hanya sekadar hasil dari usaha intelektual di dalam kelas, melainkan akumulasi dari berbagai aspek kehidupan mahasiswa. Faktor-faktor seperti latar belakang keluarga, kondisi ekonomi, kebiasaan belajar, dan karakteristik pribadi berperan krusial dalam membentuk capaian nilai akhir. Memahami faktor-faktor ini secara komprehensif memungkinkan identifikasi dini terhadap mahasiswa yang membutuhkan perhatian khusus.

Institusi pendidikan dapat memanfaatkan hasil analisis data ini untuk mengevaluasi dan meningkatkan kualitas pembelajaran. Dengan mengetahui pola yang mendasari kesuksesan atau kegagalan mahasiswa, universitas dapat menyusun strategi intervensi yang lebih proaktif, seperti program bimbingan akademik yang disesuaikan atau penyesuaian kurikulum. Tujuan utama penggunaan *machine learning* dalam konteks ini adalah memprediksi performa akademik secara akurat, sehingga tindakan preventif dapat diambil sebelum mahasiswa mengalami kegagalan, pada akhirnya meningkatkan angka kelulusan dan reputasi institusi pendidikan tersebut.

**Permasalahan yang Ingin Diselesaikan:**
"Bagaimana memprediksi performa akademik mahasiswa (kategori nilai/grade akhir) berdasarkan faktor pribadi, kondisi keluarga, dan kebiasaan belajar mereka sehari-hari?"

**Tujuan Penelitian:**

**Manfaat Penelitian:**

## 6. Data Preparation

Tahapan persiapan data dilakukan pada **KNIME Analytics Platform** untuk memastikan kebersihan data yang akan dimasukkan ke model prediksi.

### Import Dataset
Proses import dataset dilakukan dengan menggunakan node **CSV Reader**. Node ini membaca file `DATA (1).csv` ke dalam *table format* yang terstruktur. Konfigurasi dilakukan untuk memastikan separator koma (`,`) digunakan dan *header* kolom berhasil diidentifikasi.

### Pemeriksaan Data
- **Struktur Data:** Dicek menggunakan *table view* bahwa semua 145 baris dan 33 kolom sudah terbaca dengan sempurna.
- **Tipe Atribut:** Semua prediktor berupa tipe numerik *integer*.
- **Missing Value:** Dataset ini sempurna dan bebas dari nilai rumpang (0 *missing values*), sehingga tidak memerlukan intervensi *Missing Value Node*.

### Transformasi Data
Secara khusus, pada dataset asli ini, seluruh indikator kategorikal telah berbentuk *encoded numerical* (direpresentasikan dalam integer). Hal ini membuat **tidak memerlukan tahapan *encoding* tambahan** seperti *One to Many* (One-Hot Encoding) ataupun pengubahan string menjadi *number*.

### Normalisasi Data
Untuk menjaga stabilitas rentang (*range*) data pada setiap fiturnya, digunakan node **Normalizer**.
- **Fungsi:** Mengubah semua atribut menjadi berada dalam skala min-max yang sama (contoh dari rentang 0 ke 1).
- **Mengapa Normalisasi Diperlukan:** Tahapan ini diperlukan sebelum klasifikasi untuk mencegah algoritma memprioritaskan fitur dengan nilai magnitudo dominan akibat perbedaan satuan atau rentang asal. Skala yang terstandarisasi membuat laju pembentukan *tree* berbobot seragam dan adil.

## 7. Modeling

### Metode yang Digunakan
Metode penambangan data yang digunakan adalah **Decision Tree Classification**.

**Konsep Decision Tree:**
*Decision Tree* (Pohon Keputusan) adalah teknik klasifikasi hierarkis. Konsep intinya meliputi:
- **Root Node (Simpul Akar):** Atribut terbaik (*most significant*) di puncak pohon yang digunakan untuk pembelahan pertama seluruh dataset.
- **Internal Node (Simpul Internal):** Pertanyaan bersyarat atau pengujian atribut lebih lanjut di bawah *Root Node*.
- **Leaf Node (Simpul Daun):** Keputusan atau kelas target prediksinya di ujung cabang (dalam hal ini `OUTPUT Grade`).

**Alasan Pemilihan Decision Tree:**
1. **Mudah Dipahami:** Model memberikan output berupa *rules* (*if-then*) yang sangat transparan dan mirip penalaran manusia.
2. **Mudah Diinterpretasikan:** Bentuk visual pohon memudahkan manajemen/dosen membaca kondisi apa saja yang membuat mahasiswa mendapat nilai tertentu.
3. **Dapat Mengidentifikasi Faktor Penting:** Secara bawaan melakukan pemilihan fitur, menaruh atribut terpenting di hirarki teratas.

### Workflow KNIME

Workflow untuk proses *training* dan *testing* dibentuk dengan rangkaian node berikut:
- **CSV Reader:** Gerbang awal untuk mengimpor dataset berformat .csv ke dalam KNIME.
- **Number to String:** Mengubah tipe data kolom target (`OUTPUT Grade`) dari numerik (Integer) menjadi kategorikal (String) untuk target klasifikasi.
- **Table Partitioner:** Membagi dataset secara acak menjadi dua bagian: Data Latih (Train 80%) dan Data Uji (Test 20%).
- **Decision Tree Learner:** Melatih data *Train* untuk membangun algoritma pohon keputusan dan memodelkan pola.
- **Decision Tree Predictor:** Menerapkan model yang telah dibuat untuk memprediksi target kelas pada Data Uji yang belum pernah dilihat model.
- **Decision Tree View:** Menampilkan visualisasi bentuk pohon keputusan (*tree*) untuk memudahkan membaca aturan (rules).
- **Scorer:** Membandingkan nilai tebakan dari node Predictor dengan nilai aslinya di Data Uji, lalu menghitung tingkat keberhasilan (akurasi) model.

## 8. Evaluation

Penggunaan **Node Scorer** dalam KNIME ditujukan untuk mengukur dan memvalidasi keakuratan model klasifikasi *Decision Tree* yang telah dilatih.

### Accuracy
**Definisi:**
Accuracy (Akurasi) adalah ukuran kedekatan nilai tebakan dari sebuah model terhadap label atau nilai aslinya secara umum di semua kelas kategori. 

**Rumus Accuracy:**
$$
\text{Accuracy} = \frac{\text{True Positive} + \text{True Negative}}{\text{Total Data Uji}}
$$

### Confusion Matrix
Evaluasi untuk detil tiap kategori (*Fail*, *DD*, dsb) diamati via matriks silang, dengan definisi elemen (jika berfokus pada salah satu target kelas positif):
- **True Positive (TP):** Prediksi mengatakan mahasiswa berada di *Grade X*, dan kenyataan sebenarnya juga di *Grade X*.
- **True Negative (TN):** Prediksi mengatakan bukan *Grade X*, dan kenyataan sebenarnya juga bukan *Grade X*.
- **False Positive (FP):** Prediksi mengatakan *Grade X*, namun aslinya salah (kesalahan menebak positif).
- **False Negative (FN):** Prediksi mengatakan bukan *Grade X*, namun aslinya justru *Grade X* (gagal mengenali kejadian).

**Interpretasi Hasil dan Perhitungan Akurasi (Berdasarkan Output Node Scorer):**

**1. Pembuktian Data Uji (Test Set)**
Jika kita menjumlahkan elemen pada salah satu baris Confusion Matrix, kita akan mendapatkan total data uji:
- **Rumus:** True Positive (TP) + False Positive (FP) + True Negative (TN) + False Negative (FN)
- **Baris 1:** 5 + 2 + 19 + 3 = **29 data**
Ini membuktikan *Table Partitioner* (rasio 80:20) telah berjalan, karena 20% dari 145 baris data adalah 29 data untuk diuji.

**2. Perhitungan Overall Accuracy (0.379)**
Akurasi keseluruhan diperoleh dari jumlah total tebakan benar (Total True Positive) dibagi total data uji.
- **Total Tebakan Benar (Semua TP):** 5 + 2 + 2 + 0 + 0 + 0 + 0 + 2 = **11**
- **Akurasi:** 11 / 29 = 0.3793... (**0.379** atau **37.9%**)
Angka 37.9% ini menunjukkan model hanya menebak benar 11 dari 29 data. Akurasi ini rendah dikarenakan dataset menargetkan klasifikasi kompleks **8 kelas nilai (multi-class)** (Fail hingga AA), di mana terdapat tumpang tindih (*overlap*) pola kebiasaan mahasiswa pada *grade* yang berdekatan.

**3. Cohen's Kappa (0.27)**
Nilai *Cohen's Kappa* sebesar 0.27 berada dalam rentang **"Fair Agreement"** (Tingkat kesesuaian wajar). Ini menegaskan bahwa meskipun akurasi 37.9%, model tidak sekadar menebak jawaban secara acak (random guess), melainkan sudah memprediksi berdasarkan *rules* algoritma pohon keputusan yang valid.

## 9. Analisis Faktor yang Mempengaruhi Nilai Mahasiswa

Berdasarkan *rule set* model Decision Tree yang didapatkan, identifikasi atribut yang paling berpengaruh kuat dapat disimpulkan. Contoh urutan atribut yang mendominasi simpangan klasifikasi:

1. **Attendance to Classes:** Sering kali menjadi simpul pembagi utama. Alasan: Keterlibatan fisik dan fokus mahasiswa secara sinkronus di dalam ruang pembelajaran merepresentasikan penyerapan dasar materi. 
2. **Weekly Study Hours:** Waktu belajar tambahan menjustifikasi dedikasi eksternal mahasiswa untuk mereviu materi, mengolah tugas, dan persiapan ujian.
3. **Reading Frequency:** Merefleksikan literasi sains dan komprehensi verbal mahasiswa. Mahasiswa yang lebih sering membaca mampu menganalisis permasalahan kompleks lebih terstruktur di ujian.
4. **Taking Notes:** Menunjukkan keseriusan mahasiswa dalam menyortir poin-poin utama materi. Hal ini berkorelasi langsung dengan kemampuan *recall* teori sebelum ujian.
5. **Previous GPA (akumulasi semester lalu, jika direpresentasikan):** Sejarah kebiasaan akademik mahasiswa seringkali linear dengan etos belajar yang telah terbangun.

**Interpretasi Akademis:**
Dari perspektif teori pendidikan, temuan ini menguatkan paradigma bahwa keaktifan dan strategi pembelajaran mahasiswa yang inisiatif (*student-centered factors*) adalah determinan primer keberhasilan studi. Interaksi konstan (*Attendance* & *Notes*) dipadukan penguatan mandiri (*Study Hours* & *Reading*) menciptakan siklus penyerapan ilmu dan *grade* optimal.

## 10. Kesimpulan

1. **Penerapan Data Mining Efektif:** Data mining, khususnya algoritma *machine learning*, dapat secara presisi digunakan untuk merancang prediksi mengenai performa akademik mahasiswa berdasarkan histori data mereka.
2. **Keandalan Decision Tree:** Model *Decision Tree* terbukti sangat kapabel untuk mengklasifikasikan multi-kelas (*OUTPUT Grade*) pada kasus mahasiswa dan menyediakan representasi visual atau aturan baku yang bisa diinterpretasi mudah.
3. **Peran Utama Kehadiran:** Kehadiran kuliah merupakan faktor prediktor kuat yang sangat berpengaruh menopang mahasiswa untuk menjauhi nilai di bawah batas kelulusan.
4. **Kunci Nilai Sempurna:** Variabel jam belajar mandiri dan kebiasaan membaca berkontribusi linear terhadap peningkatan prestasi untuk meraih indeks nilai teratas (*BA/AA*).
5. **Solusi Peringatan Dini:** Model yang dibangun dapat dimanfaatkan secara praktis sebagai dasar pengambilan keputusan akademik bagi dosen dalam upaya peringatan dini dan tindakan konseling terhadap mahasiswa rawan gagal studi.

## 11. Saran

- **Bagi Mahasiswa:** 
  Mahasiswa diharapkan dapat merefleksikan pentingnya tingkat kehadiran penuh dan perlunya mempertahankan habitus mencatat materi maupun belajar secara konsisten di luar jadwal reguler, sebagai formula keberhasilan capaian akademik.

  Dosen dan penasihat akademik disarankan menggunakan rekap jejak belajar (*tracking*) awal semester untuk segera mendekati mahasiswa yang absen berulang kali, sehingga tindakan preventif untuk menghindari drop grade dapat diberikan secepatnya.
- **Bagi Peneliti Selanjutnya:** 
  Penelitian berikutnya disarankan untuk menambah ragam metode klasifikasi (misal: SVM atau Neural Networks) sebagai pembanding akurasi, dan disarankan mengompilasi rekam jejak dataset berukuran lebih besar (lintas fakultas atau tahun angkatan yang berbeda) untuk memperkokoh akurasi *generalization* model algoritma.

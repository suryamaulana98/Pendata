# Normalisasi Data pada Data Mining

Notebook ini menjelaskan konsep **normalisasi data** sebagai bagian dari proses **data preprocessing** dalam Data Mining.

Normalisasi sangat penting ketika menggunakan algoritma yang berbasis **perhitungan jarak**, karena perbedaan skala data dapat menyebabkan atribut tertentu lebih dominan dibandingkan atribut lainnya.

## Pengertian Normalisasi Data

Normalisasi data adalah proses **mengubah nilai atribut ke dalam skala tertentu** sehingga setiap atribut memiliki kontribusi yang seimbang dalam proses analisis data.

Dalam banyak kasus pada Data Mining dan Machine Learning, dataset memiliki atribut dengan skala yang sangat berbeda.

Contoh:

- Harga Rumah → 600 – 1500 (dalam juta)
- Jumlah Kamar → 2 – 4
- Luas Tanah → 90 – 200

Jika algoritma menggunakan **perhitungan jarak seperti Euclidean Distance**, maka atribut dengan nilai yang lebih besar (misalnya harga rumah) akan memiliki pengaruh yang jauh lebih besar dibandingkan atribut lain.

Oleh karena itu diperlukan proses **normalisasi** agar semua atribut berada pada skala yang sebanding.

Algoritma yang sangat membutuhkan normalisasi antara lain:

- K‑Nearest Neighbor (KNN)
- Weighted KNN (WKNN)
- Clustering (K‑Means)
- Cosine Similarity


## Catatan Penting

Pada proses normalisasi **hanya atribut numerik yang dinormalisasi**.

Atribut kategorikal seperti:

- Nama Produk
- Jenis Rumah
- Kota
- Kategori Barang

tidak dapat langsung dinormalisasi karena nilainya **bukan angka**.

Oleh karena itu, pada proses preprocessing biasanya dilakukan:

1. Memilih atribut numerik
2. Melakukan normalisasi hanya pada atribut tersebut


## Dataset Contoh (Data Harga Rumah)

| Rumah | Luas_Tanah | Luas_Bangunan | Jumlah_Kamar | Harga |
|------|------|------|------|------|
| R1 | 120 | 90 | 3 | 850 |
| R2 | 200 | 150 | 4 | 1500 |
| R3 | 150 | 110 | 3 | 1000 |
| R4 | 90 | 70 | 2 | 600 |

Pada dataset ini terdapat beberapa atribut:

- **Rumah** → atribut kategorikal
- **Luas Tanah** → atribut numerik
- **Luas Bangunan** → atribut numerik
- **Jumlah Kamar** → atribut numerik
- **Harga** → atribut numerik

Hanya atribut **numerik** yang akan dilakukan proses normalisasi.

```python
import pandas as pd

data = {
    'Rumah':['R1','R2','R3','R4'],
    'Luas_Tanah':[120,200,150,90],
    'Luas_Bangunan':[90,150,110,70],
    'Jumlah_Kamar':[3,4,3,2],
    'Harga':[850,1500,1000,600]
}

df = pd.DataFrame(data)
df
```

## 1. Min‑Max Normalization

Min‑Max Normalization adalah metode normalisasi yang **mengubah nilai data ke rentang tertentu**, biasanya antara **0 sampai 1**.

Metode ini mempertahankan hubungan proporsional antar data.

### Rumus

$$
x' = \frac{x - x_{min}}{x_{max} - x_{min}}
$$

Keterangan:

- $x$ = nilai data asli
- $x_{min}$ = nilai minimum atribut
- $x_{max}$ = nilai maksimum atribut
- $x'$ = nilai hasil normalisasi


```python
def min_max_normalization(column):
    return (column - column.min()) / (column.max() - column.min())

minmax_df = df.copy()

for col in ['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga']:
    minmax_df[col] = min_max_normalization(df[col])

minmax_df
```

## Min‑Max dengan Rentang Baru

Min‑Max juga dapat digunakan untuk mengubah data ke **rentang lain**, misalnya 0 – 10.

### Rumus

$$
x' = \frac{x - x_{min}}{x_{max} - x_{min}} \times (new_{max} - new_{min}) + new_{min}
$$

Keterangan:

- $new_{min}$ = batas minimum baru
- $new_{max}$ = batas maksimum baru


```python
def minmax_new_range(column,new_min=0,new_max=10):
    return ((column-column.min())/(column.max()-column.min()))*(new_max-new_min)+new_min

range_df = df.copy()

for col in ['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga']:
    range_df[col] = minmax_new_range(df[col],0,10)

range_df
```

## 2. Z‑Score Normalization (Standardization)

Metode ini menormalisasi data berdasarkan **rata‑rata (mean)** dan **standar deviasi**.

Metode ini sering digunakan pada algoritma statistik.

### Rumus

$$
z = \frac{x - \mu}{\sigma}
$$

Keterangan:

- $\mu$ = mean (rata‑rata)
- $\sigma$ = standar deviasi


```python
def z_score(column):
    return (column - column.mean()) / column.std()

zscore_df = df.copy()

for col in ['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga']:
    zscore_df[col] = z_score(df[col])

zscore_df
```

## 3. Decimal Scaling Normalization

Metode ini melakukan normalisasi dengan **membagi nilai atribut dengan pangkat 10**.

### Rumus

$$
x' = \frac{x}{10^j}
$$

Keterangan:

- $j$ = jumlah digit dari nilai maksimum dalam atribut


```python
def decimal_scaling(column):
    max_val = column.abs().max()
    j = len(str(int(max_val)))
    return column/(10**j)

decimal_df = df.copy()

for col in ['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga']:
    decimal_df[col] = decimal_scaling(df[col])

decimal_df
```

## 4. Mean Normalization

Mean Normalization memusatkan data di sekitar **nilai rata‑rata**.

### Rumus

$$
x' = \frac{x - \mu}{x_{max} - x_{min}}
$$


```python
def mean_normalization(column):
    return (column-column.mean())/(column.max()-column.min())

mean_df = df.copy()

for col in ['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga']:
    mean_df[col] = mean_normalization(df[col])

mean_df
```

## Normalisasi menggunakan Scikit‑Learn

Python menyediakan library **sklearn.preprocessing** untuk melakukan normalisasi secara otomatis.


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled = scaler.fit_transform(df[['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga']])

scaled_df = pd.DataFrame(scaled,columns=['Luas_Tanah','Luas_Bangunan','Jumlah_Kamar','Harga'])

scaled_df
```

## Kesimpulan

Normalisasi data merupakan tahap penting dalam **data preprocessing** pada Data Mining.

Tujuan normalisasi adalah:

- Menyamakan skala antar atribut
- Mencegah atribut tertentu mendominasi perhitungan
- Meningkatkan performa algoritma berbasis jarak

Metode normalisasi yang umum digunakan antara lain:

- Min‑Max Normalization
- Z‑Score Normalization
- Decimal Scaling
- Mean Normalization

Dalam praktiknya, **hanya atribut numerik yang dinormalisasi**, sedangkan atribut kategorikal biasanya diproses menggunakan teknik lain seperti encoding.
